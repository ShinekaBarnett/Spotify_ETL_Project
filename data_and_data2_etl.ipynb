```python
import pandas as pd
import sqlalchemy
from sqlalchemy import create_engine
```

```python
### Extracted CSVs into DataFrames
```

```python
data_file = "Resources/data.csv"
data_df = pd.read_csv(data_file)
data_df.head()
```

```python
data2_file = "Resources/data2.csv"
data2_df = pd.read_csv(data2_file, dtype='unicode')
data2_df.head()
```

```python
### Transformed Data CSV DataFrame
```

```python
data_cols = ["id","valence", "acousticness", "artists", "energy","tempo"]
data_transformed = data_df[data_cols].copy()

data_transformed = data_transformed.rename(columns = {"id": "id","valence":"Valence", "acousticness" :"Acousticness", 
                                                        "artists":"Artists", "energy":"Energy",
                                                         "tempo":"Tempo"})

data_transformed.drop_duplicates("id", inplace = True)
data_transformed.set_index("id", inplace = True)

data_transformed.head()
```

```python
### Transform Data2 CSV DataFrame
```

```python
data2_cols = ["artist_id","valence", "acousticness", "artist_names", "liveness","tempo"]


data2_transformed = data2_df[data2_cols].copy()
data2_transformed = data2_transformed.rename(columns = {"artist_id": "id",
                                                        "valence":"Valence", 
                                                        "acousticness" :"Acousticness", 
                                                        "artist_names":"Artists", 
                                                        "liveness":"Energy",
                                                        "tempo":"Tempo"})

data2_transformed.drop_duplicates("id", inplace = True)
data2_transformed.set_index("id", inplace = True)

data2_transformed.head()
```

```python
### Created database connection
```

```python
connection_string = "postgres:postgres@localhost:5432/spotify_db"

engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm tables
engine.get_table_names()
```

```python
data_transformed.to_sql(name='data', con=engine, if_exists='append', index=True)
```

```python
data2_transformed.to_sql(name='data2', con=engine, if_exists='append', index=True)
```
